# Neuronas lineales adaptativas

Tambien conocidas como ADALINE, representan una mejora del modelo de perceptron, aplicando una diferencia muy sutil en la fase de actualizacion de pesos. En el modelo  ADALINE los pesos se actualizan en base a una funcion lineal continua en vez de a una funcion escalon unitario. La funcion de activacion $\phi(z)$ es simplemente la misma z:

$$\phi(\textbf{w}^T\textbf{x}) = \textbf{w}^T\textbf{x}$$

Mientras que la funcion lineal asi definida se utiliza para actualizar los pesos, seguiremos utilizando la funcion umbral para realizar la prediccion final, que es parecida a la del perceptron.

Notese que la comparacion de las etiquetas se realiza en base a la funcion de activacion lineal (rojo).

## Minimizar funciones de coste con el descenso del gradiente

Una de los ingredientes fundamentales en los algoritmos de aprendizaje automatico es una **funcion objetivo** que debe ser optimizada durante el proceso de aprendizaje. Esta funcion suele ser una funcion de coste que queremos minimizar y la cual crecera con la cantidad de errores que se presenten.

En el caso de ADALINE, la funcion de coste se definira como la **suma de los errores cuadraticos** (**SSE**) entre la salida calculada y la etiqueta de clase verdadera:

$$J(\textbf{w}) = \frac{1}{2}\sum_i\Big(y^{(i)} - \phi(z^{(i)})\Big)^2$$ 

El termino 1/2 no hace parte de la definicion de SSE, pero se coloca para facilidad de calculos posteriores.

La principal ventaja de esta funcion vs la funcion escalon unitario, es que esta es diferenciable, y por lo tanto que se puede optimizar para buscar los pesos que minimicen la funcion.

Para hallar los pesos que logran minimizar la funcion, nos aprovecharemos de que la funcion es convexa y por lo tanto sera cuestion de aplicar un metodo muy sencillo llamada **el descenso del gradiente**.



La idea de este metodo es bajar la colina, dando pasos en direccion contraria al gradiente y de longitud igual al rango de aprendizaje $\eta$ asi como por la pendiente del gradiente.

Con esta tecnica podemos actualizar los pesos haciendo un paso en la direccion opuesta del gradiente $\triangledown J(\textbf{w})$ de la funcion de coste:

$$ \textbf{w}:=\textbf{w} + \Delta \textbf{w}$$

en donde $\Delta \textbf{w}$ estara dado por:

$$\Delta \textbf{w}=-\eta \triangledown J(\textbf{w})$$

Para obtener el valor de $\triangledown J(\textbf{w})$, necesitamos calcular la derivacion parcial de la funcion de coste con respecto a cada peso $w_j$:

$$\frac{\partial J}{\partial w_j}=-\sum_i \Big(y^{(i)}-\phi\big(z^{(i)}\big)\Big)x_j^{(i)}$$

Por lo que la actualizacion del peso $\textbf{w}$ se  puede escribir como:

$$\Delta w_j = -\eta \frac{\partial J}{\partial w_j}=\eta \sum_i \Big(y^{(i)}-\phi\big(z^{(i)}\big)\Big)x_j^{(i)}$$

Es importante tener en cuenta que la actualizacion del peso se realiza en base a todas las muestras de entrenamiento, en lugar de actualizar los pesos de forma incremental despues de cada muestra.


## Implementando Adaline

```python
# Librerias a utilizar
# ===========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
```

```python
class AdalineGD(object):
    """ADAptive LInear NEuron classifier.

    Parametros
    ------------
    eta : float
      Rango de aprendizaje (entre 0.0 1 1.0)
    n_iter : int
      Iteracion sobre el datset de entrenamiento completo.
    random_state : int
      Semilla de numeros aleatorios


    Atributos
    -----------
    w_ : 1d-array
      Pesos de las caracteristicas.
    cost_ : list
      Suma de cuadrados de la funcion de costo en cada epoca.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """ Fit de datos de entrenamiento.

        Parametros
        ----------
        X : {array-like}, shape = [n_samples, n_features]
        y : array-like, shape = [n_samples]

        Returno
        -------
        self : object
        """
        
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size = 1 + X.shape[1]) #shape: (filas, columnas)
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X) # m filas, 1 columna
            output = self.activation(net_input) 
            errors = (y - output) # m filas y una columnas
            self.w_[1:] += self.eta * X.T.dot(errors) # n filas y una columna
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        """Calcula net input (z)"""
        return np.dot(X, self.w_[1:]) + self.w_[0] # sale un vector de m filas y 1 columna

    def activation(self, X):
        """Calculo de la activacion lineal"""
        return X

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

```python
datos = pd.read_csv('iris.csv')

# Voy a trabajar con dos categorias unicamente: Iris-setosa e Iris-versicolos
datos = datos[(datos['Name'] == 'Iris-setosa') | (datos['Name'] == 'Iris-versicolor')]
X = datos.iloc[:, [0, 2]].values

# A iris-setosa le colocare la categoria 1 y a Iris-versicolor, la categoria -1
y = datos['Name'].values
y = np.where(y == 'Iris-setosa', 1, -1)
```

```python
modelo1 = AdalineGD(eta = 0.01, n_iter = 10).fit(X, y)
modelo2 = AdalineGD(eta = 0.0001, n_iter = 10).fit(X, y)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
x = np.arange(1, len(modelo.cost_) + 1)
y1 = modelo1.cost_
y2 = modelo2.cost_
ax[1].plot(x, np.log10(y1), marker = 'o')
ax[0].plot(x, y2, marker = 'o')
ax[1].set_ylabel('log(SSE)')
ax[1].set_xlabel('Epocas')
ax[0].set_ylabel('SSE')
ax[0].set_xlabel('Epocas')
ax[1].set_title('Adaline - Learning rate = 0.01')
ax[0].set_title('Adaline - Learning rate = 0.0001');
```

## Mejorar el descenso del gradiente mediante el escalado de caracteristicas

Recordar esto: EL DESCENSO DEL GRADIENTE SE BENEFICIA DEL ESCALADO DE CARACTERISTICAS.

Existen muchos otros algoritmos que veremos mas a delante que tambien se benfician del escalado de caracteristicas, y existen diferentes tecnicas de esclado. La que veremos aqui se llama **normalizacion**, la cual transforma nuestros datos a una distribucion normal estandar, lo cual ayuda al descenso del gradiente a converger mas rapidamente. 

Recordar que la distribucion normal estandar tiene media cero y desviacion estandar 1, por lo tanto nuestros valores de entrada adquiriran esas caracteristicas. Para normalizar la caracterisitca $j$, procederemos de la siguiente manera:

$$\textbf{x}_j'=\frac{\textbf{x}_j-\mu_j}{\sigma_j}$$

En donde $\textbf{x}_j$ es un vector que consta de las caracterisitcas $j$, es decir, una columna de la mstriz de entrada $\textbf{X}$; este procedimiento se debe aplicar a todas las caracteristicas.



```python
# Caracteristicas normalizadas
X_std = np.copy(X)
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```

```python
modelo = AdalineGD(n_iter=15, eta=0.01)
modelo.fit(X_std, y)

plot_decision_regions(X_std, y, classifier=modelo)
plt.title('Adaline - Gradiente Descendiente')
plt.xlabel('sepal length [normalizado]')
plt.ylabel('petal length [normalizado]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

plt.plot(range(1, len(modelo.cost_) + 1), modelo.cost_, marker='o')
plt.xlabel('Epocas')
plt.ylabel('SSE')

plt.tight_layout()
plt.show()
```

Notar que el SSE no es cero a pesar de que todas las muestras estan correctamente clasificadas.
