# Subplots

Nos permite graficas varias figuras en el mismo lienzo por medio de la creacion de varios ejes diferentes. 

```python
%matplotlib notebook

import matplotlib.pyplot as plt
```

```python
plt.subplots?
```

```python
import numpy as np
x = np.linspace(1, 10, 100)
y1 = x
y2 = x**2

fig, ax = plt.subplots(1, 2, sharey = True)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
```

```python
y3 = x**2.1
ax[1].plot(x, 10*y3)
```

```python
x = np.linspace(1, 10, 100)
y1 = x
y2 = x**2
y3 = np.cos(x)
y4 = np.sin(x)

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)
```

# Histogramas

```python
datos = np.random.randint(1, 7, 1000)
fig, ax = plt.subplots()
ax.hist(datos, bins = 6, align = 'mid', edgecolor = "r")
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

#10**(n+1)= 10, 100, 1000, and 10000 
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title(f'n={sample_size}')
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
side_histogram.hist(Y, bins=100, orientation='horizontal')
plt.show()
```

```python
top_histogram.clear()
top_histogram.hist(X, bins=100)
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal')
# flip the side histogram's x axis
side_histogram.invert_xaxis()
```

# Diagramas de caja y bigotes

```python
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample, 
                   'random': random_sample, 
                   'gamma': gamma_sample})
df.head()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots()
# create a boxplot of the normal data, assign the output to a variable to supress output
ax.boxplot(df['normal'], whis = 'range')
```

```python
plt.clf()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
plt.show()
```

```python
fig, ax = plt.subplots()
plt.boxplot([df['normal'], df['random'], df['gamma']])
plt.show()
```
