# Graficando con pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
# Creando un dataframe
array = np.random.random(60).reshape(15, 4)
df = pd.DataFrame(array, columns = ["A", "B", "C", "D"])
```

```python
df.head()
```

Grafica de lineas

```python
df.plot(kind = 'line')
```

Grafica de puntos

```python
df.plot('A', 'B', kind = "scatter");
```

Tambien es posible definir el tipo de grafico con el formato df.plot.type, usando los siguientes tipos:

    'line' : line plot (default)
    'bar' : vertical bar plot
    'barh' : horizontal bar plot
    'hist' : histogram
    'box' : boxplot
    'kde' : Kernel Density Estimation plot
    'density' : same as 'kde'
    'area' : area plot
    'pie' : pie plot
    'scatter' : scatter plot
    'hexbin' : hexbin plot

```python
df.plot.scatter("A", "B", s= 100, c='r')
```

```python
df.plot.scatter("A", "B", c = "D", s = df["D"]*100)
```

```python
df.plot.scatter("A", "B", c = "D", s = df["D"]*100, colormap = 'viridis')
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B']*100, colormap='viridis')
ax.set_aspect('equal')
```

```python
%matplotlib inline
df.plot.box();
```

```python
df.plot.hist(alpha=0.5);
```

```python
df.plot.kde();
```

# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
fig, ax = plt.subplots()
ax.hist(v1, alpha=0.7, bins=np.arange(-50,150,5), label='v1');
ax.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2');
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.hist([v1, v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1,v2))
sns.kdeplot(v3);
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

```python
sns.jointplot(v1, v2, alpha=0.4)
```

```python
grid = sns.jointplot(v1, v2, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(v1, v2, kind='hex');
```

```python
sns.set_style('white')

sns.jointplot(v1, v2, kind='kde', space=0);
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.pairplot(iris, hue='Name', diag_kind='kde', height=1.5);
```

```python
plt.figure(figsize=(8,6))
plt.subplot(121)
sns.swarmplot('Name', 'SepalLength', data=iris);
plt.subplot(122)
sns.violinplot('Name', 'SepalLength', data=iris);
```
